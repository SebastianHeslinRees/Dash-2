```python
#function for convert values to percentages
columns = region_mirco_census_sample.columns

def convert_values_to_percentages(region_mirco_census_sample, col):
    p = pd.DataFrame(region_mirco_census_sample.groupby(['region_id',col]).size())
    p_pcts = p.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
    df_pcts = pd.pivot_table(p_pcts, values=0, index=['region_id'],
                        columns=[col], aggfunc=np.sum, fill_value=0).round(2)
    return df_pcts

DFs = []
for col in columns: #lsit of cols
    print("col: "+str(col))
    df_pcts = convert_values_to_percentages(region_mirco_census_sample, col) #generate df
    DFs.append(df_pcts)
df_concated = pd.concat(DFs, axis=0, ignore_index=True)

from functools import reduce
df = reduce(lambda df1,df2: pd.merge(df1,df2,on='region_id'), DFs)
```

```python
def get_highlights(selections):
    geojson_highlights = dict()
    for k in regions.keys():
        if k != 'features':
            geojson_highlights[k] = regions[k]
        else:
            geojson_highlights[k] = [lookup[selection] for selection in selections]        
    return geojson_highlights
```

```python
def get_figure(selections):
    # Base map layer
    fig = px.choropleth_mapbox(df, locations="region_id", hover_data=dict({'White: English/Welsh/Scottish/Northern Irish/British2':False, 'region_id':False}), color_discrete_map={
                        'Missing Data': '#FFFFFF',
                        '20-39' : '#77DE5F',
                        '60-79' : '#4A84BA',
                        '80-100' : '#314FA8'}, category_orders={
                      'White: English/Welsh/Scottish/Northern Irish/British2' : [
                          'Missing Data',
                          '20-39',
                          '60-79',
                          '80-100'
                      ]
                    }, featureidkey="properties.rgn19cd", 
                            geojson=regions, color="White: English/Welsh/Scottish/Northern Irish/British2", opacity=0.5)

    # Second layer
    if len(selections) > 0:
        # highlights contain the geojson information for only selected authority
        
        highlights = get_highlights(selections)
        for i in range(df['White: English/Welsh/Scottish/Northern Irish/British2'].nunique()):
            
            fig.add_trace(
                px.choropleth_mapbox(df, geojson=highlights, 
                                 color="White: English/Welsh/Scottish/Northern Irish/British2",
                                 locations="region_id", color_discrete_map={
                      'Missing Data': '#FFFFFF',                        
                        '20-39' : '#77DE5F',                       
                        '60-79' : '#4A84BA',
                        '80-100' : '#314FA8'}, category_orders={
                      'White: English/Welsh/Scottish/Northern Irish/British2' : [
                          'Missing Data',                          
                          '20-39',                        
                          '60-79',
                          '80-100'
                      ]
                    }, hover_data=dict({'White: English/Welsh/Scottish/Northern Irish/British2':False, 'region_id':False}), hover_name=None,
                                 featureidkey="properties.rgn19cd",                              
                                 opacity=1).data[i]
                )

    #------------------------------------#   
    fig.update_traces(hoverinfo=None)
    fig.update_layout(mapbox_style="carto-positron", showlegend=False,
                      mapbox_zoom=6, height=900,
                      mapbox_center={"lat": 55, "lon": 0},
                      margin={"r":0,"t":0,"l":0,"b":0},
                      uirevision='constant')
    
    return fig
```

```python
def get_card(selections):
    row = df[df['region_id'] == selections]
    row = row.reset_index(drop = True)
    if selections == '':
        card_content1a = []
        return card_content1a
    
    name = row['Area'][0]
    
    if row['White: English/Welsh/Scottish/Northern Irish/British2'][0] == 'Missing Data':
        card_content1a = [
            dbc.CardBody([
                html.Div(html.H2("{}".format(name))),
                html.Br(),
                html.Div(html.H4("Data missing."))])]
        
        return card_content1a
    
    rate = row['White: English/Welsh/Scottish/Northern Irish/British'][0]

    
    card_content1a = [
        dbc.CardBody(
            [
                html.Div(html.H2("{}".format(name))),
                html.Br(),
                html.Div(html.H4("Percentage Ethnicity")),
                html.Div([html.H3(["{}".format(rate)])]), 
                ]
            ),
        ]   
    
    return card_content1a
```

```python
badges = html.Div(
    [   
        dbc.Badge("Missing Data", color="light"),
        dbc.Badge("20-39", color="#77DE5F"),
        dbc.Badge("60-79", color="#4A84BA"),
        dbc.Badge("80-100", color="#314FA8"),   
        
    ]
)
```

#########################################################################################################

```python
@app.callback(
    [Output('choropleth', 'figure'),
     Output('info', 'children')],
    [Input('choropleth', 'clickData')])
def update_figure(clickData):
    selections = set()
    
    if clickData is not None:            
        location = clickData['points'][0]['location']
        selections.add(location)
        return get_figure([location]), get_card(location)
                    
    return get_figure([]), get_card('')
```
